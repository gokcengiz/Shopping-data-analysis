```python
df.groupby("shopping_mall").describe().T
```

In Istanbul, the most shopping is done at Kanyon AVM, and the least shopping is done at Mall of Istanbul. The average age of customers shopping at shopping malls is generally the same(43). Based on the average, the most spending was made at Emaar Square Mall.

```python
df.groupby("gender").describe().T
```

If we take the shopping customers as a basis according to their gender; The age statistics of male and female shoppers are almost the same. Women have done more shopping, and if we look at the average of spending, men have spent more.

```python
df.groupby("payment_method").describe().T
```

44447 customers completed their purchases with cash, 34931 customers with a credit card and 20079 customers with a debit card. If we take the average expenditure as a basis, cash expenditures were relatively higher.

```python
df.groupby("category").describe().T
```

Shopping expenses were made mostly for clothes and least for books.
